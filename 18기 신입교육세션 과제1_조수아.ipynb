1. 데이터명 : YouTube Trending Videos 2025 (Updated Daily)
데이터 특성
출처 : Kaggle (YouTube API 기반 Trending Videos Dataset)
(https://www.kaggle.com/datasets/sebastianbesinski/youtube-trending-videos-2025-updated-daily)
분야 : 소셜미디어 / 콘텐츠 트렌드 분석
내용 : 이 데이터는 YouTube API를 통해 수집된 전 세계 인기 급상승(Trending) 동영상 정보를 포함하고 있으며, 90개 이상의 국가에서 매일 업데이트되는 데이터셋이다. 영상 ID, 제목, 채널명, 국가, 조회 수, 좋아요 수, 댓글 수, 업로드 날짜, 수집 날짜 등의 변수를 포함하고 있다. 이를 통해 국가별 인기 콘텐츠 비교, 조회 수와 사용자 반응 지표(좋아요·댓글) 간의 관계 분석, 시점에 따른 트렌드 변화 분석이 가능하다. 특히 한국(KR)을 포함한 다양한 국가 데이터를 활용하여 글로벌 콘텐츠 트렌드와 지역별 선호도 차이를 분석하는 데 적합한 데이터이다.

2. 데이터명 : Survey Data on Financial Literacy and Asset Participation among Young Adults in Indonesia (2025)
데이터 특성
출처 : Mendeley Data (https://data.mendeley.com/datasets/k269sxcw6h/1)
분야 : 금융 문해력 / 투자 행동 분석 
내용 : 이 데이터는 2025년에 수집된 청년 성인의 금융 이해력과 자산 참여(저축·주식·펀드 참여 등) 관련 설문 응답을 포함한다. 기본적·고급 금융 지식 수준, 주식 및 뮤추얼펀드 참여 여부, 인구통계 변수(성별·소득 등), 저축 행태 등이 포함돼 있다. 이를 통해 금융 문해력 수준이 실제 투자 선택 및 자산 참여에 어떤 영향을 미치는지 분석할 수 있다. 특히 최근 주식 및 자산 투자 열풍과 금융 지식 격차의 관계를 데이터 기반으로 살펴볼 수 있는 자료이다.
